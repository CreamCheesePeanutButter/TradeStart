| Field | Meaning                                      |
| ----- | -------------------------------------------- |
| `T`   | ticker symbol                                |
| `v`   | trading **volume** (number of shares traded) |
| `vw`  | **volume-weighted average price**            |
| `o`   | **open price** (price when market opened)    |
